The evolution of the density matrix is given by
\begin{equation}
\frac{d\rho}{dt} = -iL_0\rho-iL_{int}\rho-i\Gamma\rho
\end{equation}
For numerical and perturbative solutions, it is convenient to define the density matrix in the interaction picture as
\begin{equation}
\rho_{I}(t) = \exp(i(L_0+\Gamma)t/\hbar)\rho(t=0)\exp(-i(L_0+\Gamma)t/\hbar),
\end{equation}
and the perturbing operator in the interaction picture,
\begin{equation}
L_I(t) = \exp(i(L_0+\Gamma)t/\hbar)\rho(t=0)\exp(-i(L_0+\Gamma)t/\hbar).
\end{equation}
The evolution of the density matrix in the interaction picture is then given by
\begin{equation}
i\hbar\frac{d}{dt}\rho_I(t) = L_I(t)\rho_I(t).
\end{equation}

## Valence Manifold of States Coupled to Core-Excited Manifold of States

We describe the system with the following Hamiltonian.
\begin{equation}
H = H_0+H_{int}.
\end{equation}
Here, $H_0$ is the Hamiltonian of an atom without an applied X-ray field,
\begin{equation}
H_0 = \sum_{v\in V, c\in C}|v>\epsilon_v<v|+|c>\epsilon_c<c|,
\end{equation}
where $V$ denotes the set of valence states, and $C$ denotes the set of core-excited states. $\epsilon_v$ is the energy of state $|v>$. $H_{int}$ is the Hamiltonian due to interaction with the X-ray field,
\begin{equation}
H_{int} = -\mu E(t),
\end{equation}
with $E(t)$ the electric field of the applied X-rays, and $\mu$ the electric dipole operator,
\begin{equation}
\mu = \sum_{v\in V, c\in C}|v>\mu_{vc}<c|+|c>\mu_{cv}<v|.
\end{equation}
Note that we neglect dipolar coupling between valence states and between core-excited states.

## Coupling of Material to X-Rays

We write the applied electric field and material polarization as products of a slowly varying envelope function and a rapidly oscillating carrier wave,
\begin{equation}
E(z, t) = \mathcal{E}(z, t)\exp(i(kz-\omega t))+c.c.
\end{equation}
\begin{equation}
P(z, t) = \mathcal{P}(z, t)\exp(i(kz-\omega t))+c.c.
\end{equation}

The interaction Hamiltonian in the interaction picture is then given by
\begin{equation}
H^I = \exp(iH_0 t/\hbar)H_{int}\exp(-iH_0t/\hbar)
\end{equation}
\begin{equation}
H^I = \left( \sum_{v\in V, c\in C} |v>\mu_{vc}<c|\exp(-i\omega_{cv}t)+|c>\mu_{cv}<v|\exp(i\omega_{cv}t) \right) \left(\mathcal{E}(t)\exp(-i\omega t)+\mathcal{E}^*(t)\exp(i\omega t)\right).
\end{equation}
Now, we expect the density matrix elements in the interaction picture to vary slowly compared to the carrier frequency ω. The above expression for $HˆI$ has two types of terms. First, there are terms which vary slowly compared to the carrier frequency, then there are terms which oscillate with frequencies around twice the carrier frequency. The terms which oscillate rapidly do not drive changes in the density matrix effectively. We now make the Rotating Wave Approximation, which neglects these rapidly oscillating terms. The interaction Hamiltonian in the interaction picture within this Rotating Wave Approximation is
\begin{equation}
H^I = \sum_{v\in V, c\in C} \mathcal{E}^*(t)|v>\mu_{vc}<c|\exp(i(\omega-\omega_{cv})t)+\mathcal{E}(t)|j>\mu_{cv}<v|\exp(-i(\omega-\omega_{cv})t)
\end{equation}
We can simplify the above expression by defining the complex time-dependent Rabi frequency for the transition between valence state |v> and core-excited state |c> as
\begin{equation}
R_{vc}(t) = \mathcal{E}^*(t) \mu_{vc}\exp(i(\omega-\omega_{cv})t)/\hbar = R_{cv}^*.
\end{equation}
$H^I$ can then be written as 
\begin{equation}
H^I = \hbar \sum_{v\in V, c\in C} |v>R_{vc}<c|+|c>R_{cv}<v|.
\end{equation}


and the evolution of the density matrix in the interaction picture is
\begin{equation}
i\hbar \frac{\partial \rho^I}{\partial t} = L^I\rho^I-\Gamma \rho^I,
\end{equation}
with
\begin{equation}
L^I = (H^I \otimes I)-(I\otimes (H^I)^T)
\end{equation}

## Calculation of spectrum exiting sample

Once we know the density matrix in the interaction picture at a given moment, we can solve for the slowly-varying polarization envelope,
\begin{equation}
\mathcal{P}(t) = \sum_{i, j}\mu_{ij}\rho_{ji}^I\exp(i(\omega_{ij}+\omega)t)
\end{equation}
The envelope of the X-rays exiting the sample is
\begin{equation}
\mathcal{E}(t) = \mathcal{E}_{in}(t)+i\frac{2\pi\omega}{c}\mathcal{P}(t)
\end{equation}

## Summary of Simulation Procedure

### 1. Define an Electric Field for the X-ray Pulse Incident on the Sample
We assume an electric field given by
\begin{equation}
E(z, t) = \mathcal{E}(z, t)\exp(i(kz-\omega t))+c.c.
\end{equation}
For a Gaussian envelope electric field, the envelope is given by
\begin{equation}
\mathcal{E}(z, t) = a\exp\left(\frac{-4\ln(2)(t-t_0)^2}{w^2}\right),
\end{equation}
where $w$ is the full-width at half-maximum.

### 2. Define System
Material starts in ground state:
\begin{equation}
\rho(t = -\infty) = |g> \otimes |g>.
\end{equation}
Define the energies of the states in the absence of an applied electromagnetic field:
\begin{equation}
H_0 = \sum_s |s> \epsilon_s <s|.
\end{equation}
Define the dipole matrix elements between valence states and core-excited states:
\begin{equation}
\mu = \sum_{v\in C, c\in C} |v> \mu_{vc}<c|+|c>\mu_{cv}<v|
\end{equation}
Define the relaxation superoperator $\Gamma$.

### 3. Calculate the Time Dependence of the Density Matrix in the Interaction Picture
The interaction Hamiltonian in the interaction picture is given within the rotating wave approximation by
\begin{equation}
H^I \approx \hbar \sum_{v\in V, c\in C} |v>R_{vc}<c|+|c>R_{cv}<v|,
\end{equation}
where
\begin{equation}
R_{vc}(t) = \mathcal{E}^*(t)\mu_{vc}\exp(i(\omega-\omega_{cv})t)/\hbar = R_{cv}^*(t).
\end{equation}
The corresponding operator in Liouville space is
\begin{equation}
L^I = (H^I \otimes I)-(I \otimes (H^I)^T).
\end{equation}
The evolution of the density matrix in the interaction picture is then given by
\begin{equation}
i\hbar \frac{\partial \rho^I}{\partial t} = L^I\rho^I-\Gamma\rho^I.
\end{equation}
Thus,
\begin{equation}
\rho^I(t-t_0) = \exp\left[-\frac{i(L^I-\Gamma)(t-t_0)}{\hbar}\right]\rho^I(t_0)
\end{equation}

### 4. Calculate the Time-Dependent Material Polarization
The time-dependent polarization is given by
\begin{equation}
\mathcal{P}(z, t) = \sum_{i, j}\mu_{ij}\rho_{ji}^I\exp(i(\omega_{ij}+\omega)t)
\end{equation}

### 5. Calculate the Electric Field of the X-Rays Exiting the Sample
The envelope of the electric field of the X-rays exiting the sample is
\begin{equation}
\mathcal{E}(t) = \mathcal{E}_{in}(t)+i\frac{2\pi\omega}{c}\mathcal{P}(t)
\end{equation}
